```python
from __future__ import division, print_function
import numpy as np
import matplotlib.pyplot as plt
from concurrent.futures import ProcessPoolExecutor
import scipy.stats

import weighted_photon_tools as wpt
```

```python
def fakeprobs(n, minprob=1e-3):
    return np.exp(np.random.uniform(np.log(minprob),0,size=n))
def fake_uniform(n):
    return np.random.uniform(-0.5,0.5,size=n)
def fake_vm(probs, kappa):
    r = fake_uniform(len(probs))
    c = np.random.uniform(0,1, size=len(probs)) < probs
    vm = np.random.vonmises(mu=0, kappa=kappa, size=np.sum(c))/(2*np.pi)
    r[c] = vm
    return r
```

```python
def true_vm(kappa, bins):
    xs = np.linspace(-np.pi, np.pi, bins+1)
    c = scipy.stats.vonmises(kappa).cdf
    return (c(xs[1:])-c(xs[:-1]))


kappa = 10
bins = 20
print(np.sum(true_vm(kappa, bins)))
vmd = np.random.vonmises(mu=0, kappa=kappa, size=1000)/(2*np.pi)
plt.plot(np.histogram(vmd, bins=bins, range=(-0.5, 0.5))[0],drawstyle="steps-mid")
plt.plot(len(vmd)*true_vm(kappa,bins),drawstyle="steps-mid")
```

```python
kappa = 10
bins = 40
probs = fakeprobs(100000)
phases = fake_vm(probs, kappa)

bg = wpt.background_level(probs)/bins
sf = wpt.foreground_scale(probs)
print(sf, (1-wpt.background_level(probs)/np.sum(probs))**(-1))

bv, be, _ = plt.hist(phases, weights=probs, bins=bins, histtype='step', range=(-0.5,0.5))
print(np.sum(bv-bg), np.sum(probs)-bins*bg)
plt.plot(np.repeat(be,2)[1:-1], bg+np.sum(probs)*(sf**(-1))*np.repeat(true_vm(kappa, bins),2))
plt.xlim(-0.5,0.5)
```

```python
def check_h(n):
    probs = fakeprobs(n)
    phases = fake_uniform(n)
    assert len(probs)==len(phases)
    return wpt.h(phases,probs)
e = ProcessPoolExecutor()
f_hs = []
```

```python
check_h(10000)
```

```python
for i in range(10000):
    f_hs.append(e.submit(check_h, 10000))
```

```python
hs = [f.result() for f in f_hs if f.done()]
print("%d/%d" % (len(hs),len(f_hs)))
hs.sort()
xs = np.log10((np.arange(len(hs))+1)/len(hs))
plt.plot(hs,xs)
plt.plot(xs,xs)
```

```python
bins = 8
predicted_p = (wpt.background_level(probs)/bins
               +np.sum(probs)*true_vm(kappa, bins)/wpt.foreground_scale(probs))
def check_histogram_errors(probs,kappa):
    photons = fake_vm(probs, kappa)
    p, pe, be = wpt.weighted_histogram_errors(photons,probs, bins=bins)
    zs = (p-predicted_p)/pe
    return zs
e2 = ProcessPoolExecutor()
f_zs = []
```

```python
check_histogram_errors(probs, kappa)
```

```python
for i in range(10000):
    f_zs.append(e2.submit(check_histogram_errors, probs, kappa))
```

```python
zs = [f.result() for f in f_zs if f.done()]
print("%d/%d" % (len(zs),len(f_zs)))
zs = np.array(zs)
zflat = zs.flat
lcdfs = scipy.stats.norm.logcdf(zflat)
lsfs = scipy.stats.norm.logsf(zflat)
lcdfs.sort()
lsfs.sort()
xs = np.log((np.arange(len(lsfs))+1)/len(lsfs))
plt.plot(lcdfs,xs, label="lower tail")
plt.plot(lsfs,xs, label="upper tail")
plt.plot(xs,xs)
plt.legend(loc='best')
```

```python
time = 1000 # days, say
rate = 1. # per day
kappa = 10. # concentration in phase

total_probability = time*rate
n = 10000
while True:
    probs = fakeprobs(n)**4
    cprobs = np.cumsum(probs)
    i = np.searchsorted(cprobs, total_probability)
    if i<len(cprobs):
        probs = probs[:i+1]
        break
    else:
        n *= 2
print(len(probs),np.sum(probs))
phases = fake_vm(probs, kappa=kappa)
```

```python
H, bins = wpt.plot_result(phases,probs,time=time)

be = np.linspace(-0.5,0.5,bins+1)
tvm = bins*true_vm(kappa, bins)
print(np.mean(tvm))
#sf = np.sum(probs)/np.sum(probs**2)
#sf2 = (np.sum(probs)-np.sum(probs*(1-probs)))/np.sum(probs)
#print(sf, sf2)
plt.plot(np.repeat(be,2)[1:-1], np.repeat(rate*tvm,2))
```
